```python
def split_sequences(sequences, n_steps):
  X, y = list(), list()
  for i in range(len(sequences)):
    # find the end of this pattern
    end_ix = i + n_steps
    # check if we are beyond the dataset
    if end_ix > len(sequences) - 1:
      break
    # gather input and output parts of the pattern
    seq_x, seq_y = sequences[i:end_ix, -1], sequences[end_ix, -1]
    seq_x = np.append(seq_x, sequences[end_ix, :-1])
    X.append(seq_x)
    y.append(seq_y)
  return array(X), array(y)
```

```python
train_data = train_df.loc[:, ['Holiday', 'Discount', 'Sales']].values
```

```python
stores = sorted(train_df['Store_id'].unique())
```

```python
def get_models_dict(model, n_steps):

  def train_model(model, X, y):
    local_model = clone(model)
    local_model.fit(X, y)
    return local_model

  models = {}
  for store in stores: 
    print('Training model for Store ', store, '...', end='')
    data = train_df[train_df['Store_id'] == store]
    data = data.loc[:, ['Holiday', 'Discount', 'Sales']].values
    X, y = split_sequences(data, n_steps)
    local_model = train_model(model, X, y)
    models[store] = local_model
    print('done')

  return models
```

```python
from collections import OrderedDict
```

```python
def make_predictions(models, n_steps):
  predictions = {}

  for store in stores: 
    print('Making predictions for Store ', store , '...', end='')
    data = test_df[test_df['Store_id'] == store]
    ids = data['ID'].values
    data = data.loc[:, ['Holiday', 'Discount']].values
    prev_sales = train_df[train_df['Store_id'] == store].iloc[-n_steps: , :]
    prev_sales = prev_sales.values[:, -1]
    for i in range(len(data)):
      fv = np.append(prev_sales, data[i,:])
      fv = fv.reshape((1, len(fv)))
      pred = models[store].predict(fv)[0]
      predictions[ids[i]] = pred
      prev_sales = np.delete(prev_sales, [0])
      prev_sales = np.append(prev_sales, pred)
    print('done')

  ordered_ids = test_df['ID'].values
  ordered_predictions = {}

  for id in ordered_ids:
    ordered_predictions[id] = predictions[id]

  return ordered_predictions
```
